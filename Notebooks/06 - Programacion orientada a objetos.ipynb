Python es un lenguaje orientado a objetos y todo lo que incluye son en efecto objetos, pero se puede utilizar sin definir clases tranquilamente como lenguaje de scripting o definir estructuras, variables de manera independiente, mientras la sintaxis sea correcta, funcionará.

1. Agrupar datos y su comportamiento asociado en un mismo lugar

Comencemos definiendo una clase "Alumno":

```python
class Alumno:
  def __init__(self, nombre, apellido, fecha_nac, legajo):
    self.nombre = nombre
    self.apellido = apellido
    self.fecha_nacimiento = fecha_nac
    self.legajo = legajo
```

Instanciemos un objeto de tipo Alumno:

```python
alumno_1 = Alumno('Manuel', 'Quintana', '1990-01-17', 1052201)
print(type(alumno_1))
```

Veamos los atributos de alumno_1

```python
print(alumno_1.nombre)
print(alumno_1.apellido)
print(alumno_1.fecha_nacimiento)
print(alumno_1.legajo)
```

Ahora agreguemos métodos a la clase Alumno, por ejemplo un método que permita devolver nombre completo.

```python
class Alumno:
  def __init__(self, nombre, apellido, fecha_nac, legajo):
    self.nombre = nombre
    self.apellido = apellido
    self.fecha_nacimiento = fecha_nac
    self.legajo = legajo

  def get_nombre_completo(self):
    nombre_completo = self.nombre + ' '  + self.apellido
    return nombre_completo
```

```python
#Volvemos a instanciar un alumno de clase Alumno
alumno_1 = Alumno('Manuel', 'Quintana', '1990-01-17', 1052201)
print(f'El nombre completo es: {alumno_1.get_nombre_completo()}')
```

Ahora vamos a ver a nivel teórico otros conceptos de OOP

**1. Herencia**

* La herencia habilita a una clase a heredar métodos y atributos de una o más clases base.

* La definición se hace una sola vez en lugar de tenerla repetida en varios lados (Code organization)

**2. Herencia múltiple**

* Con herencia múltiple, una clase podrá heredar de más de una clase base.

* Por la complejidad extra de esto, no se usa mucho al desarrollar pero es útil al desarrollar **interfaces**.



**3. Interfaces**

* Una interfaz es una promesa/contrato que hace una clase, que asegurará cierto comportamiento o capacidad en cada clase que herede de ella.

* Es una combinación de herencia múltiple y clases abstractas.

* Python no trae built-in interfaces, hay que desarrollarlas utilizando abstract base classes y herencia múltiple.

* Ej: Método que comparten varias clases pero que tiene una implementación distinta según sobre qué instancia se ejecute



**4. Composición**

* Consiste en crear objetos complejos a partir de objetos más simples.
