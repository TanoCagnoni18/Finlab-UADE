# Estructuras de datos

Según la documentación oficial Python ya trae 4 estructuras de datos incorporadas que funcionan como "colecciones de datos" y que podemos usar de acuerdo a nuestras necesidades.

Estas son:

- lista: `lista = [elemento 1, elemento2, elemento3]`

- tupla: `tupla = (elemento1, elemento2, elemento3)`

- set: `set = {elemento1, elemento2, elemento3}`

- diccionario: `dict = {'clave1': valor1, 'clave2': valor2, 'clave3':valor3}`

¿Y cuales son las diferencias? Vemos que hay dos categorías bien separables:

**1. Estructuras ordenadas**:

- List is a collection which is ordered and changeable. Allows duplicate members
- Tuple is a collection which is ordered and unchangeable. Allows duplicate members

**2. Estructuras desordenadas**:

- Set is a collection which is unordered and unindexed. No duplicate members.
- Dictionary is a collection which is unordered and changeable. No duplicate members.

Vamos a estudiar en detalle las listas y los diccionarios que son las estructuras que más a menudo vamos a utilizar.

Así mismo hay otras estructuras de datos, que no son propias de Python y que van a aparecer por ejemplo al momento de querer almacenar información o al momento de querer publicar información que será consumida por un programa. Más adelante las presentaré cuando nos toque trabajar con ellas.

# Listas

```
'''
Listas

En lugar de tener por ejemplo 5 variables:
numero1 = 1
numero2 = 1
numero3 = 1412421
numero4 = 5
numero5 = 13
print(numero1)
print(numero2)
print(numero3)
print(numero4)
print(numero5)

podemos trabajar con una sola lista que tenga 5 numeros
'''
lista = [1,2,4,542,25]
print('Muestro toda la estructura de la lista:')
print(lista)

# Una lista en python es un objeto iterable
print('Forma Pythonic: Muestro los elementos de la lista iterando 1 a 1:')
for i in lista:
  print(i)

print('Forma clásica de acceder a los elementos de una lista')
for i in range(0,5):
  print(lista[i])
```
