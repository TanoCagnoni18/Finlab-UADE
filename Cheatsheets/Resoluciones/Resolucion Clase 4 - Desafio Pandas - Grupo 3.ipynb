```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```
#### 2) Importar los datos dentro de un objeto Python / Numpy / Pandas para su manipulación

def cargaDatacsv(path,delimiter,skip=None,erase_comment=None):#Armo funcion para cargar los datos paso 3 parametros path(ruta del archivo), delimeter(que me delimita los rows) & skip (la cantidad de filas que quiero saltear).
  df = pd.read_csv(path, sep = delimiter,skiprows=skip) #Elimino las 11 primeras lineas
  return df

#seteo variables a usar en la creacion del DF inicial.
path='https://raw.githubusercontent.com/al34n1x/DataScience/master/3.Pandas/udemy_courses_ejercicio.csv'
delimiter=';'
rows_to_skip=11

df=cargaDatacsv(path,delimiter,rows_to_skip)
df.drop(df.index[0],inplace=True) #Elimino la linea basura
df.head()
```

```
#### 3) Identificar nulos 
df.isnull().sum()
```

```
### reemplazarlos por valores acordes
df.fillna({'level':'NoData','price':0}, inplace=True)#completamos los NaN dependiendo el datatype
df['content_duration'].round()
df.head(10)
```

```
#### 4) Reemplazar Nombre de columnas por nuevos descriptivos
df.rename(columns = {'course_id':'id','course_title':'titulo','is_paid':'es_pago','price':'precio','num_subscribers':'num_suscriptores','num_reviews':'num_resenas','num_lectures':'num_lecturas','level':'nivel','content_duration':'duracion_contenido','published_timestamp':'marca_tiempo','subject':'asunto'}, inplace = True)
df.head(10000)
```

```
##5) Ordenar / Rankear los cursos por una variable/columna que nos proveea la relevancia del mismo
df.sort_values(by=['num_suscriptores'])#ordenamos por numero de suscriptores
```

```
 ##Crear al menos dos DF diferentes con información que considere relevante
df_top_all=df[df['nivel']== 'All Levels'].sort_values('num_suscriptores',ascending=False).head(10)#Generamos varios DF para cada nivel de los cursos
df_top_int=df[df['nivel']== 'Intermediate Level'].sort_values('num_suscriptores',ascending=False).head(10)
df_top_beg=df[df['nivel']== 'Beginner Level'].sort_values('num_suscriptores',ascending=False).head(10)
df_top_exp=df[df['nivel']== 'Expert Level'].sort_values('num_suscriptores',ascending=False).head(10)


df_top_all.reset_index(drop=True,inplace=True)
df_top_int.reset_index(drop=True,inplace=True)
df_top_beg.reset_index(drop=True,inplace=True)
df_top_exp.reset_index(drop=True,inplace=True)
```

```
YSusc_all = df_top_all['num_suscriptores']
XSusc_all = df_top_all['titulo'].str[:30]
YSusc_int = df_top_int['num_suscriptores']
XSusc_int = df_top_int['titulo'].str[:30]
YSusc_beg = df_top_beg['num_suscriptores']
XSusc_beg = df_top_beg['titulo'].str[:30]
YSusc_exp = df_top_exp['num_suscriptores']
XSusc_exp = df_top_exp['titulo'].str[:30]

fig, ax = plt.subplots(2, 2, figsize=(18,12), sharex=True)
plt.subplots_adjust(wspace=0.5, hspace=0.2)

ax[0,0].barh(XSusc_all, YSusc_all, color="tab:orange")
ax[0,0].set_title('Top 10 Nivel = All Levels', fontsize=14, fontweight='bold')
ax[0,1].barh(XSusc_beg, YSusc_beg, color="tab:blue")
ax[0,1].set_title('Top 10 Nivel = Beginner Level', fontsize=14, fontweight='bold')
ax[1,0].barh(XSusc_int, YSusc_int, color="tab:red")
ax[1,0].set_title('Top 10 Nivel = Intermediate Level', fontsize=14, fontweight='bold')
ax[1,1].barh(XSusc_exp, YSusc_exp, color="tab:green")
ax[1,1].set_title('Top 10 Nivel = Expert Level', fontsize=14, fontweight='bold')

plt.xlabel('Cantidad de Suscriptores', fontsize=12, fontweight='bold')
plt.show()
```

```
promedio = df.groupby('asunto')['precio'].mean().round()

fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(promedio.index,promedio,label='Promedio',color="tab:blue",)

ax.set_title('Costo Promedio de Cursos por Categoria', loc = "Center", fontdict = {'fontsize':16, 'fontweight':'bold', 'color':'tab:blue'})
ax.legend(loc = 'upper left')
plt.xlabel("Categorias de Estudio",fontsize=20)
plt.ylabel("Costo Promedio",fontsize=20)
plt.show
```

```
# Discretizacion de los datos por el valor de los cursos para observar la distribucion de los mismos
bins=[0,50,100,150,200]

PreciosArray=pd.value_counts(pd.cut(df['precio'].tolist(),bins)) # Se debe transformar en Lista para que luego el objeto pueda ser un Pandas Array
                                                                 # y poder aplicar el metodo value_counts

rangos=['0-50','50-100','100-150','150-200']

fig, ax = plt.subplots(figsize=(10, 7))
ax.bar(rangos,PreciosArray,label='Cursos',color="tab:orange")


ax.set_title('Distribucion de Cursos x Precios', loc = "Center", fontdict = {'fontsize':18, 'fontweight':'bold', 'color':'tab:blue'})
ax.legend(loc = 'upper right')
plt.xlabel("Rangos de Precios",fontsize=15)
plt.ylabel("Cantidad de Cursos",fontsize=15)
plt.show
```

```
df_top_cursosxsus=df.sort_values('num_suscriptores').tail(15)
x_grafico=df_top_cursosxsus['num_suscriptores']
y_grafico=df_top_cursosxsus['titulo']
fig, ax = plt.subplots(figsize=(10, 7))

ax.barh(y_grafico,x_grafico,label='Titulos',color="tab:green",)

ax.set_title('Top 15 - Cursos con mayor cantidad de suscriptores', loc = "Center", fontdict = {'fontsize':16, 'fontweight':'bold', 'color':'tab:blue'})
ax.legend(loc = 'lower right')
plt.xlabel("Cantidad de Suscriptores",fontsize=20)
plt.ylabel("Titulos",fontsize=20)
plt.show
```
