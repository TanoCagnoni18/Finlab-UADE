# Parte 1: Estructura condicional


**Ejercicio 1**

Ingresar dos números enteros e indicar si son iguales o distintos.



---


**Ejercicio 2**

Ingresar las notas de los dos parciales de un alumno e indicar si promocionó, aprobó o si debe recuperar. Informar del error si alguna nota no está entre 0 y 10.
Criterios:
1. Se promociona cuando ambas notas son mayores o iguales a 7.
2. Se aprueba cuando ambas notas son mayores o iguales a 4 pero alguna es menor a 7.
3. Se recupera cuando al menos una de las dos notas está por debajo de 4.

---
**Ejercicio 3**

Solicitar el ingreso de un número entero y mostrar un mensaje indicando si es par o impar.



---

**Ejercicio 4**

Una remisería requiere un programa que calcule el precio del viaje a partir de la cantidad de km recorridos. La tarifa es la siguiente:
1. Viaje mínimo \$100
2. Si recorre entre 0 y 10 km: \$20/km
3. Si recorre 10 km o más: \$15/km

# Parte 2: Estructura iterativa

**Ejercicio 5**

Realizar un programa que muestre en pantalla la tabla del 7, desde 1 hasta 10 con el siguiente formato:

7 x 1 = 7

7 x 2 = 14

...

7 x 10 = 70


---
**Ejercicio 6**

Desarrollar un programa que calcule y muestre la suma de todos los números impares comprendidos entre 42 y 278.








---
**Ejercicio 7**

Ingresar números por teclado hasta que la suma de los mismos supere el número 1000. Una vez finalizado, mostrar el resultado de la suma.



---

**Ejercicio 8**

Leer 10 números enteros ingresados por el usuario e informa cual fue el mayor valor leído.




---
**Ejercicio 9**

Desarrollar un programa que calcule la facturación diaria de un negocio.

El usuario ingresa uno a uno los montos de cada factura emitida durante el día. El final del ingreso de datos se realiza cuando el usuario ingresa el monto 0.

Mostrar al finalizar, la facturación total del día.
