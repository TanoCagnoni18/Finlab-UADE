**Ejercicio 1**

Diseñar una función que devuelva el máximo entre tres números recibidos como parámetros.



---


**Ejercicio 2**

Programar una función que reciba una lista de números como parámetro y calcule la suma de los números.

---
**Ejercicio 3**

Programar una función que reciba una lista con al menos 5 números como parámetro y retorne los 3 menores.



---
**Ejercicio 4**

Escribir una función signo(n) que devuelva 1, 0 o -1 según el valor recibido como parámetro sea positivo, nulo o negativo.


---
**Ejercicio 5**

Escribir una función que reciba como parámetro dos números enteros y devuelva la multiplicación de los mismos utilizando únicamente sumas.

Ej: 4 * 3 = 4 + 4 + 4

---
**Ejercicio 6**

**Ejercicio Desafio**

Escribir una función que reciba como parámetro una lista (puede ser de números o de caracteres alfanuméricos) y determine si la misma es capicúa devolviendo 1 en caso que lo sea y -1 en caso que no lo sea.
